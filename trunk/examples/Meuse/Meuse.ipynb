```python
from osm2hydro.utils.gdal_readmap import gdal_readmap
from osm2hydro.utils.basemap_imshow import basemap_imshow
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *

minx = 5.1
miny = 50.3
maxx = 5.9
maxy = 50.9

fig = figure(figsize=(9,8))

def show_grid(grid_filename):
    carray = np.array([0.3,0.4,0.5,0.6,0.7,0.8,0.9,1.0])
    cmap = cm.jet
    norm = mpl.colors.BoundaryNorm(carray,cmap.N)
    x, y, grid, zz = gdal_readmap(grid_filename, 'PCRaster')
    filt = grid < carray.min()
    grid[filt] = np.nan

    return basemap_imshow(x, np.flipud(y), np.flipud(grid), minx, maxx, miny, maxy, norm=norm, 
               meridian_dist=0.2, parallel_dist=0.2, resolution='h')

subplot(2,2,1)
show_grid('pavedcorinaresamp.map')
title('Corine',fontsize=11)

subplot(2,2,2)
show_grid('lu_pavedpol.map')
title('OSM Polygons',fontsize=11)

subplot(2,2,3)
show_grid('lu_paved_fromroads.map')
title('OSM Roads',fontsize=11)

subplot(2,2,4)
m, im = show_grid('lu_paved.map')
title('OSM combined',fontsize=11)

m.drawmapscale(5.3, 50.38, 5.3, 50.38, 20, barstyle='fancy')
axx = fig.add_axes([0.15,0.03,0.7,0.02])
fig.colorbar(im,cax=axx,orientation='horizontal',extend='min')

# savefig('liege-four.pdf',dpi=300)
# savefig('liege-four.png',dpi=300)
```
