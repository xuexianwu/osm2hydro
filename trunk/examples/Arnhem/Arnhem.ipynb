## Overview

This example demonstrates how osm2hydro can be used. The OSM files for the globe or selected areas can be downloaded here: http://wiki.openstreetmap.org/wiki/Planet.osm

The following functionality is included:

* Crop large osm file
* Run osm2hydro and visualize
    * Convert osm file into a number of Shapefiles (see osm2shp.ini)
    * Convert resulting Shapefiles into a raster file
* Analyze results
    * Compare results to the landuse maps (Corine, GBKN)

## Show results

```python
# define function used to draw shapefile (move to utils)

xmin=5.89
xmax=6.02
ymin=51.955
ymax=52.03

def plot_shapefile(ax, path, c='b',a=1.0):
    import ogr
    import os
    shp = ogr.Open(path)

    l = shp.GetLayer(0)
    t = l.GetGeomType()

    verts = []
    if t == ogr.wkbPolygon:
        for f in shp.GetLayer(0):
            g = f.GetGeometryRef()
            g = g.GetGeometryRef(0)
            points = g.GetPoints()
            verts.append(points)
        polygons = matplotlib.collections.PolyCollection(verts, edgecolors='none')
        polygons.set_facecolor(c)
        polygons.set_alpha(a)
    elif t == ogr.wkbLineString:
        for f in shp.GetLayer(0):
            g = f.GetGeometryRef()
            points = g.GetPoints()
            verts.append(points)
        polygons = matplotlib.collections.PolyCollection(verts, edgecolors=c, closed=False)
        polygons.set_facecolor('w')
    else:
        raise Exception("Support for geometry type %d is not implemented yet" %fg.GetGeometryType())
        
    ax.add_collection(polygons)

    ax.ticklabel_format(style = 'sci', useOffset=False)
    ax.set_title(os.path.basename(path) + '(' + str(l.GetFeatureCount()) + ')')
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

```
